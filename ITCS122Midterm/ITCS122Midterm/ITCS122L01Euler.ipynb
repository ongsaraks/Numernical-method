```python
import numpy as np

def euler(x0,xf,y0,dx):
  """
  Input:
    initial condition = (x0,y0), step size = dx
    xf, the value of x at where we wanà¸° to approximate
  Output:
    values of x and y at each iteration
  """
  idx = 0
  x,y = [x0],[y0]
  while True:
    x.append(x[idx] + dx)
    y.append(y[idx] + deriv(x[idx], y[idx]) * dx)
    print(deriv(x[idx], y[idx]))
    idx = idx+1
    if x[idx] >= xf-1e-9: break
  return x,y

deriv = lambda x,y: x*y #dy/dx = 2y
x,y = euler(0,0.4,1,0.1) #initial condition
print('x=',x)
print('y=',y)
```

```python
import matplotlib.pyplot as plt

xx = np.linspace(0,3,21)
yy = np.exp(2*xx)

plt.plot(x,y, 'bo--', label='approximate')
plt.plot(xx,yy, 'g', label='exact')
plt.title('Euler''s method approximates ODE')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend(loc='upper left')
plt.show()
```
